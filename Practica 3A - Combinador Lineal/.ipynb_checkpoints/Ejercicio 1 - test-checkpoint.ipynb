```python
from sympy import symbols, diff

# establece variables a usar en la funcion
x, y, z= symbols('x y z', real=True)
```

```python
def dx(x, y):
    dx_1 = (-2 * (x**2)) + ( 3 * (y**2)) + 1
    dx_2 = ( (2*(x**2)) + (3*(y**2)) + 1 )**2
    return dx_1 / dx_2
    
    
def dy(x, y):
    dy_1 = -6*x*y
    dy_2 = ( (2*(x**2)) + (3*(y**2)) + 1 )**2
    return dy_1 / dy_2
```

a) Indique el vector gradiente para  = 1 e  = 1

```python
x = 1
y = -1
print('El vector gradiente para x={} y={} {}'.format(x, y, [dx(x, y), dy(x, y)]))
```
