# Ejercicio 3b

Calcule el m√°ximo de la funci√≥n ùëì(ùë•, ùë¶) del ejercicio 1) utilizando en todos los casos MAX_ITE = 100 y una cota de 10ùëí ‚àí 06 para el valor absoluto de la diferencia de dos valores consecutivos de la funci√≥n.

Indique cu√°les de las siguientes configuraciones logra encontrar la mejor soluci√≥n (donde la funci√≥n tome el mayor valor). 

i. Punto de inicio: (ùë• = ‚àí‚àö0.5, ùë¶ = 0) ; alfa = 0.5

ii. Punto de inicio: (ùë• = ‚àí0.5, ùë¶ = 2.0) ; alfa = 0.5

iii. Punto de inicio: (ùë• = 1.5, ùë¶ = 1.5) ; alfa = 0.05

```python
from funciones import meshgridXY, graficar3D, optimize_function
import math
```

```python

def f(x, y):
    return x / ((2*(x**2)) + (3*(y**2)) + 1)

def dx(x, y):
    dx_1 = (-2 * (x**2)) + ( 3 * (y**2)) + 1
    dx_2 = ( (2*(x**2)) + (3*(y**2)) + 1 )**2
    return dx_1 / dx_2
    
    
def dy(x, y):
    dy_1 = -6*x*y
    dy_2 = ( (2*(x**2)) + (3*(y**2)) + 1 )**2
    return dy_1 / dy_2
    
X, Y = meshgridXY()
Z = f(X, Y)

graficar3D(X, Y, Z)
```

```python
alfa_value = 0.5
MAX_ITE_value = 100
cota_error_value = 1e-6
optimizer = optimize_function(dx, dy, f)

x_initial, y_initial = -1*math.sqrt(1.5), 0
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)

x_initial, y_initial = -0.5, 2.0
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)

x_initial, y_initial = 1.5, 1.5
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)
```
