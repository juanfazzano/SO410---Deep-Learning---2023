# Ejercicio 4

El precio de una vivienda, expresado en miles de d√≥lares, est√° dado por dos factores: la proximidad a centros
comerciales, financieros, v√≠as y dem√°s que se mide a trav√©s del factor j; y de la condici√≥n misma de la
vivienda que se mide por el factor k. La relaci√≥n entre estos factores y el precio de la vivienda, en un barrio
determinado, est√° dada por:

ùëÉ(ùëó, ùëò) = ùëò^2 ‚àí 2ùëóùëò + 2ùëó^2 + 18ùëó ‚àí 20ùëò + 201

Utilice la t√©cnica de descenso de gradiente para determinar los valores de j y k que minimizan el precio de
la vivienda.

```python
from sympy import symbols, diff

j, k= symbols('j k', real=True) 

f = ùëò**2 - 2*ùëó*ùëò + 2*(ùëó**2) + 18*ùëó - 20*ùëò + 201

# x=j & y=k

def df(j, k):
    return ùëò**2 - 2*ùëó*ùëò + 2*(ùëó**2) + 18*ùëó - 20*ùëò + 201


def df_dj(pj, pk):
    df_dj = diff(f, j)
    return df_dj.evalf(subs = { j:pj, k:pk } )


def df_dk(pj, pk):
    df_dk = diff(f, k)
    return df_dk.evalf(subs = { j:pj, k:pk } )

f
```

```python
from funciones import meshgridXY, graficar3D, optimize_function
X, Y = meshgridXY()
Z = df(X, Y)

graficar3D(X, Y, Z)
```

```python
MAX_ITE_value = 100
cota_error_value = 10e-5
optimizer = optimize_function(df_dj, df_dk, df)

alfa_value = 0.2
x_initial, y_initial = 78, 16
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)

alfa_value = 0.3
x_initial, y_initial = -6, 200
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)

alfa_value = 0.05
x_initial, y_initial = 2, 5
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)
```
