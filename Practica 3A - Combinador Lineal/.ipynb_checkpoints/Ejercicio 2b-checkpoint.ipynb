```python
import numpy as np
import math
import matplotlib.pyplot as plt
from pylab import meshgrid
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from funciones import meshgridXY, graficar3D, optimize_function
import math
```

```python
def fun(x, y):
    return -3/(x**2 + y**2 + 1)

def dx_fun(x, y): # derivada parcial respecto de r
    return 6*x / ((x**2 + y**2 + 1)**2)

def dy_fun(x, y): # derivada parcial respecto de h
    return 6*y / ((x**2 + y**2 + 1)**2)
    
X, Y = meshgridXY()
Z = fun(X, Y)

graficar3D(X, Y, Z)

```

```python
# Uso de la función optimize_function
optimizer = optimize_function(dx_fun, dy_fun, fun)
x_initial, y_initial = 1.5, 1.5
alfa_value = 0.05
MAX_ITE_value = 100
cota_error_value = 1e-6
optimizer(x_initial, y_initial, alfa_value, MAX_ITE_value, cota_error_value)
```
