En referencia a la función 𝑉(𝑟, ℎ) del ejercicio 2.a), utilizando en todos los casos MAX_ITE = 100 y una cota de 10𝑒 − 06 para el valor absoluto de la diferencia de dos valores consecutivos de la función, indique cuáles de las siguientes configuraciones logra encontrar una ubicación donde 𝑉(𝑟, ℎ) vale
