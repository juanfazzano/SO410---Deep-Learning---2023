Se entrenó una red neuronal multiperceptrón para resolver un problema de clasificación y al medir su
desempeño sobre el conjunto de datos de entrenamiento se obtuvo la siguiente matriz de confusión:
     
        17 0 1 0 1
        0 12 0 0 0
        0 0 12 0 0
        2 0 0 38 0
        0 8 0 0 61

a) En base a esta información, indique:
 1. Cuántos ejemplos se utilizaron en el entrenamiento.
 2. Cuántas clases puede reconocer este multiperceptrón.
 3. Cuál es la precisión (accuracy) de la red sobre el conjunto de ejemplos completo.
 4. Cuáles son los valores de precisión de la red al responder por cada uno de los valores de clase (precision).
 5. Cuáles son los valores de sensibilidad de la red al responder por cada uno de los valores de clase (recall).

```python
import numpy as np

matriz_de_confusion = [
    [17, 0, 1, 0, 1],
    [0, 12, 0, 0, 0],
    [0, 0, 12, 0, 0],
    [2, 0, 0, 38, 0],
    [0, 8, 0, 0, 61]
]
matriz_de_confusion = np.array(matriz_de_confusion)
```

```python
cant_ej_entrenamiento = np.sum(matriz_de_confusion)
print(f"1. Cuántos ejemplos se utilizaron en el entrenamiento = {cant_ej_entrenamiento}")
```

```python
cant_clases = matriz_de_confusion.shape[1]
print(f"2. Cuántas clases puede reconocer este multiperceptrón = {cant_clases}")
```

```python
# precision
accuracy = np.sum(np.diag(matriz_de_confusion)) / np.sum(matriz_de_confusion)
print("3. Cuál es la precisión (accuracy) de la red sobre el conjunto de ejemplos completo = {:.2%}".format(accuracy))
```

```python
# calcular la precisión por clase 
## (TP + FP) para esa clase
print("4. Cuáles son los valores de precisión de la red al responder por cada uno de los valores de clase (precision).")
precisiones_por_clase = np.diag(matriz_de_confusion) / np.sum(matriz_de_confusion, axis=0)
for clase, precision in enumerate(precisiones_por_clase):
    print(f"\t Clase {clase}: {precision:.2%}")
```

```python
# calcular la sensibilidad por clase
## La sensibilidad para una clase específica se calcula dividiendo TP para esa clase entre la suma de TP + FN para esa clase
print("5. Cuáles son los valores de sensibilidad de la red al responder por cada uno de los valores de clase (recall).")
sensibilidades_por_clase = np.diag(matriz_de_confusion) / np.sum(matriz_de_confusion, axis=1)
for clase, sensibilidad in enumerate(sensibilidades):
    print(f"\t Clase {clase}: {sensibilidad:.2%}")
```

$$
F1 Score =
2 \cdot \frac{p_{\text{recisión}} \cdot \text{recal}}{p_{\text{recisión}} + \text{recal}}\
$$

```python
f1_scores = 2 * (precisiones_por_clase * sensibilidades_por_clase) / (precisiones_por_clase + sensibilidades_por_clase)
f1_scores
```
