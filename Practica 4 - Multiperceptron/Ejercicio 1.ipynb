a) En base a esta informacin, indique:
 1. Cuntos ejemplos se utilizaron en el entrenamiento.
 2. Cuntas clases puede reconocer este multiperceptrn.
 3. Cul es la precisin (accuracy) de la red sobre el conjunto de ejemplos completo.
 4. Cules son los valores de precisin de la red al responder por cada uno de los valores de clase (precision).
 5. Cules son los valores de sensibilidad de la red al responder por cada uno de los valores de clase (recall).

```python
import numpy as np

matriz_de_confusion = [
    [17, 0, 1, 0, 1],
    [0, 12, 0, 0, 0],
    [0, 0, 12, 0, 0],
    [2, 0, 0, 38, 0],
    [0, 8, 0, 0, 61]
]
matriz_de_confusion = np.array(matriz_de_confusion)
```

```python
cant_ej_entrenamiento = np.sum(matriz_de_confusion)
print(f"1. Cuntos ejemplos se utilizaron en el entrenamiento = {cant_ej_entrenamiento}")
```

```python
cant_clases = matriz_de_confusion.shape[1]
print(f"2. Cuntas clases puede reconocer este multiperceptrn = {cant_clases}")
```

```python
# precision
accuracy = np.sum(np.diag(matriz_de_confusion)) / np.sum(matriz_de_confusion)
print("3. Cul es la precisin (accuracy) de la red sobre el conjunto de ejemplos completo = {:.2%}".format(accuracy))
```

```python
# calcular la precisin por clase 
## (TP + FP) para esa clase
print("4. Cules son los valores de precisin de la red al responder por cada uno de los valores de clase (precision).")
precisiones_por_clase = np.diag(matriz_de_confusion) / np.sum(matriz_de_confusion, axis=0)
for clase, precision in enumerate(precisiones_por_clase):
    print(f"\t Clase {clase}: {precision:.2%}")
```

```python
# calcular la sensibilidad por clase
## La sensibilidad para una clase especfica se calcula dividiendo TP para esa clase entre la suma de TP + FN para esa clase
print("5. Cules son los valores de sensibilidad de la red al responder por cada uno de los valores de clase (recall).")
sensibilidades_por_clase = np.diag(matriz_de_confusion) / np.sum(matriz_de_confusion, axis=1)
for clase, sensibilidad in enumerate(sensibilidades):
    print(f"\t Clase {clase}: {sensibilidad:.2%}")
```

```python
f1_scores = 2 * (precisiones_por_clase * sensibilidades_por_clase) / (precisiones_por_clase + sensibilidades_por_clase)
f1_scores
```
