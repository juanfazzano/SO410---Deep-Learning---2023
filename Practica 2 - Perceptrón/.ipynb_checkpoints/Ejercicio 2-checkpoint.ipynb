#### a) Utilice los ejemplos del archivo hojas.csv para entrenar un perceptrón que permita reconocer cuando se trata de una hoja de helecho. Utilice una velocidad de aprendizaje (parámetro alfa) de 0.01 y una máxima cantidad de iteraciones MAX_ITE=300.

```python
import numpy as np

def activacion(W, X, b):
    # W = pesos
    # X = datos de entrada
    # b = umbral

    # w1*x1 + w2*x2 + ⋯ + wn*xn
    suma_ponderada = (W * X).sum() + b
    return 1 if suma_ponderada > 0 else 0

# importar data
data = np.genfromtxt('../Datos/hojas.csv', delimiter=',', dtype=None, names=True, encoding=None)

# X = Datos de entrada
X = [[row['Perimetro'], row['Area']] for row in data]

# Valores esperados
T = list(np.where(data['Clase'] == 'Helecho', 1, 0))

# Peso
W = np.random.uniform(-1, 1, size=2)
# Umbral
b = np.random.uniform(-1, 1)
# Cant repeticiones
MAX_ITE = 200
# Velocidad de aprendizaje
alfa = 0.01

for _ in range(MAX_ITE):
    error_total = 0
    for i in range(len(data)):
        prediccion = activacion(W, X[i], b)
        error = T[i] - prediccion
        error_total += error**2
        W[0] += alfa * X[i][0] * error
        W[1] += alfa * X[i][1] * error
        b += alfa * error 
    print(error_total, end=" ")
print('')
print('W1' , W[0])
print('W2' , W[1])
print('b' , b)

# activacion(pesos, [0.5, 1], b)
```

#### b) A partir de los pesos del perceptrón entrenado, indique cuál es la función discriminante obtenida.

```python
funcion_discriminante = 0
for i in X:
    funcion_discriminante += i[0] * W[0]
    funcion_discriminante += i[1] * W[1]
funcion_discriminante
```
