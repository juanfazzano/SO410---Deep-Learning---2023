### A partir de los ejemplos del archivo Automobile-simple.csv, utilizado en la práctica 1, se entrenará un combinador lineal para predecir el precio del auto (atributo 'price') en función del tamaño del motor (atributo 'engine-size'):

```python
import pandas as pd
import numpy as np
import chardet
from sklearn import model_selection,preprocessing

def open_file(path):
    # open the path with automatic encoding
    with open(path, 'rb') as file:
        encoding = chardet.detect(file.read())
    
    return pd.read_csv(path, encoding=encoding['encoding'])

data = open_file("../Datos/automobile-simple.csv")
```

### a) Antes de comenzar con el entrenamiento, complete los datos faltantes del atributo ‘price’ utilizando el valor promedio

```python
data = data[['engine-size', 'price']]

### completamos los nulos de price con el promedio
avg_price = data['price'].mean()
data['price'].fillna(avg_price, inplace=True)
for i in data['price']:
    print(i)
```

### b) Luego de completar según lo indicado en a), realice el entrenamiento del combinador lineal:

i. Sin normalizar los atributos.

ii. Normalizando ambos atributos linealmente.

iii. Normalizando ambos atributos utilizando media y desvío.

### Analice el impacto de la normalización en el entrenamiento. ¿Qué resultados obtuvo sin normalizar los atributos? ¿Alguna de las normalizaciones es mejor? Explique.

```python
def normalizacion(aux, X):
    if aux==1: 
        normalizador= preprocessing.MinMaxScaler()
    elif aux==2:
        normalizador= preprocessing.StandardScaler()
    else:
        return X
    X = normalizador.fit_transform(X)
    return X
    
NORMALIZACION = {0:'Sin Normalizar',
                1: 'Lineal Uniforme',
                2: 'Media y Desvio'}
alpha = 0.01 
MAX_ITE = 500
```

```python
TABLA = [["Normalización", "w", "b","ECM"]]
x = data['engine-size'].values
y = data['price'].values

for N in NORMALIZACION:
    # Inicializa los parámetros del modelo (w y b)
    w = 0
    b = 0
    
    x = data['engine-size'].values
    y = data['price'].values

    x = x.reshape(-1, 1)  # Reshape x a una matriz 2D
    y = y.reshape(-1, 1)  # Reshape y a una matriz 2D

    x = normalizacion(N, x)
    y = normalizacion(N, y)

    # Descenso de gradiente estocástico
    for ite in range(MAX_ITE):
        for i in range(len(x)):
            # Selecciona un ejemplo de entrenamiento al azar
            random_index = np.random.randint(len(x))
            xi = x[random_index]
            yi = y[random_index]

            # print(xi, yi)
            
            # Predicción del modelo
            y_pred = w * xi + b
            
            # Cálculo del error
            error = yi - y_pred
            
            # Actualiza los parámetros utilizando el gradiente
            if xi != 0:
                w += alpha * error * xi
                b += alpha * error


    # Calcula el MSE
    ECM = np.mean((y - (w * x + b))**2)       
    TABLA.append([NORMALIZACION[N], w, b, ECM])

from tabulate import tabulate
table = tabulate(TABLA, headers="firstrow", tablefmt="grid")
print(table)
```
