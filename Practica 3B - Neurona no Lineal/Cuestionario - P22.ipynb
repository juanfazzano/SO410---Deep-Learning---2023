```python
import pandas as pd
import chardet
import numpy as np
def open_file(path):
    # open the path with automatic encoding
    with open(path, 'rb') as file:
        encoding = chardet.detect(file.read())
    
    return pd.read_csv(path, encoding=encoding['encoding'])
df = open_file("../Datos/FrutasTrain.csv")

df
```
